```python
import numpy as np
import matplotlib.pyplot as plt

# Load the potential energy data from the .xvg file
def load_xvg(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()
        data_lines = [line for line in lines if not line.startswith(('#', '@'))]
    data = np.array([list(map(float, line.split())) for line in data_lines])
    return data

# Example usage
potential_energy_file = 'potential.xvg'  # Path to your potential energy .xvg file
data = load_xvg(potential_energy_file)

time = data[:, 0]                 # Assuming first column is time
potential_energy = data[:, 1]     # Assuming second column is potential energy

# Define your energy minimization energy value (for example, a constant)
# This should come from your energy minimization results, possibly from .mdp or other output.
energy_minimization = np.full_like(time, -200.0)  # Example constant value (replace with actual values)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(time, potential_energy, label='Potential Energy', color='blue')
plt.plot(time, energy_minimization, label='Energy Minimization', color='red', linestyle='--')
plt.title('Potential Energy vs Energy Minimization')
plt.xlabel('Time (ps)')
plt.ylabel('Energy (kJ/mol)')
plt.grid()
plt.legend()
plt.show()

# Save the plot (optional)
plt.savefig('potential_energy_vs_minimization.png')
```
