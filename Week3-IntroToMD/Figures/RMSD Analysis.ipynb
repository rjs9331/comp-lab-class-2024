```python
import numpy as np
import matplotlib.pyplot as plt

# Function to load RMSD data from an .xvg file
def load_xvg(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()
        # Skip header lines
        data_lines = [line for line in lines if not line.startswith(('#', '@'))]
    data = np.array([list(map(float, line.split())) for line in data_lines])
    return data

# Load the RMSD data
equilibrated_data = load_xvg('rmsd.xvg')
crystal_data = load_xvg('rmsd_xtal.xvg')

# Extract time and RMSD values
time_equilibrated = equilibrated_data[:, 0]  # Time in ns
rmsd_equilibrated = equilibrated_data[:, 1]  # RMSD in nm

time_crystal = crystal_data[:, 0]  # Time in ns
rmsd_crystal = crystal_data[:, 1]  # RMSD in nm

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(time_equilibrated, rmsd_equilibrated, label='Ref:Equilibrated', color='black')
plt.plot(time_crystal, rmsd_crystal, label='Ref:Crystal', color='red')
plt.title('RMSD 2JOF, Backbone')
plt.xlabel('Time (ns)')
plt.ylabel('RMSD (nm)')
plt.grid()
plt.legend()
plt.show()


```
