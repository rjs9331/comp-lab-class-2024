```python

import numpy as np

# Load the data from the XVG file
def load_xvg(filename):
    with open(filename, 'r') as f:
        # Skip header lines
        lines = f.readlines()
        data_lines = [line for line in lines if not line.startswith(('#', '@'))]
        
    # Convert the data to a NumPy array
    data = np.array([list(map(float, line.split())) for line in data_lines])
    
    return data

# Example usage
data = load_xvg('temperature.xvg')
time = data[:, 0]   # Assuming time is in the first column
temperature = data[:, 1]  # Assuming temperature is in the second column
import matplotlib.pyplot as plt

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(time, temperature, label='2JOF', color='blue')
plt.title('Temperature 2JOF, NVT Equilibration')
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.grid()
plt.legend()
plt.show()
```
