```python
import numpy as np
import matplotlib.pyplot as plt

# Load all columns from the 'end_to_end.xvg' file
data_end_to_end = np.loadtxt('end_to_end.xvg', comments=['#', '@'])

# The first column is time (in ps), and the rest are measurements
time_end = data_end_to_end[:, 0]
measurements = data_end_to_end[:, 1:]  # All measurements after the time column

# Plot the first measurement (e.g., End-to-End distance)
plt.figure(figsize=(8, 6))
plt.plot(time_end, measurements[:, 0], label='End-to-End Distance', color='b')  # Assuming first measurement is end-to-end
plt.xlabel('Time (ps)')
plt.ylabel('End-to-End Distance (nm)')
plt.title('End-to-End Distance vs Time')
plt.legend()
plt.grid()
plt.show()

# Plot another measurement (if you want, e.g., the second measurement in the data)
plt.figure(figsize=(8, 6))
plt.plot(time_end, measurements[:, 1], label='Radius of Gyration', color='r')  # Second measurement column
plt.xlabel('Time (ps)')
plt.ylabel('Radius of Gyration (nm)')
plt.title('Radius of Gyration vs Time')
plt.legend()
plt.grid()
plt.show()
```
