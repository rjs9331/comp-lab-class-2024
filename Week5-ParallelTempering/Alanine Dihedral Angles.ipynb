```python
import mdtraj as md
import matplotlib.pyplot as plt

# Load the trajectory
trajectory = md.load('adp.trr', top='adp.gro')

# Print the trajectory information (optional)
print(trajectory)

# Compute phi and psi dihedral angles
phi_indices, phi_angles = md.compute_phi(trajectory)
psi_indices, psi_angles = md.compute_psi(trajectory)

# Convert angles from radians to degrees for better interpretability
phi_angles_deg = phi_angles * 180 / 3.14159
psi_angles_deg = psi_angles * 180 / 3.14159

# Get the time from the trajectory (in ps)
time = trajectory.time / 1000  # Convert from ps to ns

# Plot phi vs time
plt.figure(figsize=(10, 6))
plt.plot(time, phi_angles_deg[:, 0], label='Phi Angle', color='b')
plt.xlabel('Time (ns)')
plt.ylabel('Phi (degrees)')
plt.title('Phi Angle vs Time')
plt.legend()
plt.show()

# Plot psi vs time
plt.figure(figsize=(10, 6))
plt.plot(time, psi_angles_deg[:, 0], label='Psi Angle', color='r')
plt.xlabel('Time (ns)')
plt.ylabel('Psi (degrees)')
plt.title('Psi Angle vs Time')
plt.legend()
plt.show()
```
