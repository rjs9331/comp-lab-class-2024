```python
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt

# Load the trajectory and topology
traj = md.load('traj_comp.xtc', top='adp.gro')

# Compute the phi and psi angles
phi_indices, phi_angles = md.compute_phi(traj)
psi_indices, psi_angles = md.compute_psi(traj)

# Convert angles from radians to degrees
phi_angles = np.rad2deg(phi_angles)
psi_angles = np.rad2deg(psi_angles)

# Plot phi vs time
plt.figure(figsize=(10, 5))
plt.plot(traj.time, phi_angles[:, 0], label='Phi')
plt.title('Phi Dihedral Angle vs Time')
plt.xlabel('Time (ps)')
plt.ylabel('Angle (Degrees)')
plt.grid(True)
plt.show()

# Plot psi vs time
plt.figure(figsize=(10, 5))
plt.plot(traj.time, psi_angles[:, 0], label='Psi')
plt.title('Psi Dihedral Angle vs Time')
plt.xlabel('Time (ps)')
plt.ylabel('Angle (Degrees)')
plt.grid(True)
plt.show()
```
