```python
import matplotlib.pyplot as plt
import numpy as np

# Load data from replica_temp.xvg
def load_xvg(filename):
    with open(filename, 'r') as f:
        data = []
        for line in f:
            if not line.startswith(('@', '#')):  # Skip metadata lines
                data.append([float(x) for x in line.split()])
    return np.array(data)

# Load replica_temp.xvg data
replica_temp_data = load_xvg('replica_temp.xvg')

# Every 100th frame for better visibility
time = replica_temp_data[::100, 0]  # Time steps (first column)
replica_temps = replica_temp_data[::100, 1:]  # Temperatures (subsequent columns)

# Plot replica movement across temperatures
plt.figure(figsize=(10, 6))

for replica in range(replica_temps.shape[1]):
    plt.plot(time, replica_temps[:, replica], label=f'Replica {replica + 1}')

plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.title('Replica Temperature Movement Over Time')
plt.legend(loc='best')
plt.grid(True)
plt.colorbar()
plt.show()
```
