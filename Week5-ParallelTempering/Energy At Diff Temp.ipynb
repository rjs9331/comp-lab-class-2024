```python
import numpy as np
import matplotlib.pyplot as plt

# Load the energy data from the .xvg files
def load_energy(filename):
    data = []
    with open(filename, 'r') as f:
        for line in f:
            if not line.startswith(('#', '@')):  # Skip headers
                data.append(float(line.split()[1]))  # Potential energy is in the second column
    return np.array(data)

# Load energy data for each temperature
energy_T300 = load_energy('energy_T300.xvg')
energy_T400 = load_energy('energy_T400.xvg')
energy_T600 = load_energy('energy_T600.xvg')

# Plot histograms
plt.figure(figsize=(10, 6))
plt.hist(energy_T300, bins=50, alpha=0.5, label='T=300K', color='blue')
plt.hist(energy_T400, bins=50, alpha=0.5, label='T=363K', color='green')
plt.hist(energy_T600, bins=50, alpha=0.5, label='T=440K', color='red')

# Labels and legend
plt.title('Potential Energy Histogram for Different Temperatures')
plt.xlabel('Potential Energy (kJ/mol)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
