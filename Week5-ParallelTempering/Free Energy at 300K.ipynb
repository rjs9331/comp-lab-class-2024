```python
import numpy as np
import matplotlib.pyplot as plt

# Load your phi and psi data for T=300
# Example: phi_300 and psi_300 should be your angle arrays
# phi_300 = np.loadtxt('T300/phi_angles.txt')
# psi_300 = np.loadtxt('T300/psi_angles.txt')

# Create the FES
def compute_fes(phi, psi, bins=25):
    phi = np.radians(phi.flatten())  # Convert to radians if needed
    psi = np.radians(psi.flatten())  # Convert to radians if needed
    
    hist, xedges, yedges = np.histogram2d(phi, psi, bins=bins, range=[[-np.pi, np.pi], [-np.pi, np.pi]])

    hist = np.where(hist > 0, hist, 1)  # Avoid log(0)
    F = -np.log(hist)
    F -= np.min(F)  # Normalize
    
    return F, xedges, yedges

# Example arrays for phi and psi (replace with your actual data)
phi_300 = np.random.uniform(-180, 180, 1000)  # Placeholder
psi_300 = np.random.uniform(-180, 180, 1000)  # Placeholder

F_300, xedges_300, yedges_300 = compute_fes(phi_300, psi_300)

# Set maximum free energy to 6 kT
F_300[F_300 > 6] = 6

# Plot the free energy surface for T=300
plt.figure(figsize=(8, 6))
X, Y = np.meshgrid(np.degrees(xedges_300[:-1]), np.degrees(yedges_300[:-1]))
plt.imshow(F_300.T, extent=[-180, 180, -180, 180], origin='lower', cmap='viridis', aspect='auto')
plt.colorbar(label="Free Energy (kT)")
plt.title("Free Energy Surface at T=300")
plt.xlabel("$\phi$ (degrees)")
plt.ylabel("$\psi$ (degrees)")
plt.show()
```
