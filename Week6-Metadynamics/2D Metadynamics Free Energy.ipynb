```python
import numpy as np
import matplotlib.pyplot as plt

# Load FES data (assumed to be in a grid format)
# Make sure 'fes_2D.dat' contains three columns: phi, psi, and FES values
fes_data = np.loadtxt('fes_2D.dat')  # Adjust this depending on your file format

# Assuming the first two columns are X (phi) and Y (psi), and the third column is Z (FES values)
X = fes_data[:, 0]
Y = fes_data[:, 1]
Z = fes_data[:, 2]

# Create a meshgrid for contour plotting
phi_unique = np.unique(X)
psi_unique = np.unique(Y)
X_grid, Y_grid = np.meshgrid(phi_unique, psi_unique)
Z_grid = Z.reshape(len(psi_unique), len(phi_unique))  # Reshape Z for contour plotting

# Create the plot
plt.figure(figsize=(8, 6))
plt.imshow(Z_grid, extent=[X.min(), X.max(), Y.min(), Y.max()], origin='lower', cmap='viridis', aspect='auto')

# Add contour lines
contours = plt.contour(X_grid, Y_grid, Z_grid, levels=15, colors='white', linewidths=0.5)
plt.clabel(contours, inline=True, fontsize=8)

# Label the axes
plt.xlabel(r'$\phi$ / radians')
plt.ylabel(r'$\psi$ / radians')

# Add a color bar to indicate free energy values
plt.colorbar(label='Free Energy (kJ/mol)')

# Add a title
plt.title('2D Free Energy Surface of Metadynamics (phi vs psi)')

plt.tight_layout()
plt.show()
```
