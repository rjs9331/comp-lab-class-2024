```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the data, skip comment lines, and specify column names
data = pd.read_csv('COLVAR', delim_whitespace=True, comment='#', names=['time', 'phi', 'psi', 'metad.bias'])

# Step 2: Filter data to show only between 0 and 2 ns
filtered_data = data[(data['time'] >= 0) & (data['time'] <= 2)]
time = filtered_data['time']
phi = filtered_data['phi']

# Step 3: Create the plot
plt.figure(figsize=(8, 6))
plt.plot(time, phi, color='blue', label=r'$\phi$')

# Labeling the axes
plt.xlabel('Time / ns')
plt.ylabel(r'$\phi$ / radians')

# Add a title (optional)
plt.title('Time evolution of the metadynamics CV (0-2 ns)')

# Customize x-axis ticks to show every 0.25 ns between 0 and 2 ns
plt.xticks(np.arange(0, 2.25, 0.25))

# Add a legend
plt.legend()

# Add a grid (optional)
plt.grid(True)

# Show the plot
plt.show()
```
