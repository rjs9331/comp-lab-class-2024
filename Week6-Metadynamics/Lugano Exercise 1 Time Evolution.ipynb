```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the HILLS file by filtering out lines starting with '#!'
hills_file = 'bck.0.HILLS'  # Update with your actual file path
with open(hills_file, 'r') as f:
    lines = f.readlines()

# Filter out lines that start with '#!'
filtered_lines = [line for line in lines if not line.startswith('#!')]

# Create a temporary DataFrame from the filtered lines
from io import StringIO
filtered_data = StringIO("".join(filtered_lines))

# Load the data into pandas DataFrame, assuming space-separated values
hills_data = pd.read_csv(filtered_data, delim_whitespace=True, names=['time', 'phi', 'sigma_phi', 'height', 'biasf'])

# Step 2: Filter data for plotting the time vs Gaussian hill height
time = hills_data['time']
hill_height = hills_data['height']

# Step 3: Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(time, hill_height, color='blue', s=5)  # 's' controls the size of the points

# Labeling the axes
plt.xlabel('Time / ns')
plt.ylabel('Gaussian hill height / kJ mol$^{-1}$')

# Set the y-axis limit to be between 0 and 1.4 kJ/mol
plt.ylim(0, 1.4)

# Customize x-axis ticks
plt.xticks(np.arange(0, 11, 2))

# Add a grid (optional)
plt.grid(True)

# Add a title (optional)
plt.title('Time evolution of the Gaussian height')

# Show the plot
plt.show()
```
